```python
def addVoice(voice_name, voice_file):
    api_endpoint = 'https://api.elevenlabs.io/v1/voices/add'
    headers = {'xi-api-key': my_api_key}
    data = {"name": voice_name, "files": [readVoiceFile(voice_file)]}
    response = requests.post(api_endpoint, headers=headers, json=data)
    if response.status_code == 200:
        json_response = json.loads(response.text)
        return json_response['voice_id']
    else:
        print("Error: voice not added")
        return None
```

```python
addVoice("Bill_test", "resource/audio/Bill_voice.mp3")
```

```python
def readVoiceFile(file_path):
    with open(file_path, 'rb') as f:
        return f.read()
```
